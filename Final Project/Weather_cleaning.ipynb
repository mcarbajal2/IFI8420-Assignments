```python
import pandas as pd
import numpy as np
```

```python
df_data = pd.read_excel("exWeather_PV_data.xlsx", sheet_name="data")
df_data.head()
```

```python
df_data.info()
```

```python
#Change the 0 value to NaN
df_data.loc[df_data['PR'] == 0, 'PR'] = np.nan

#Remove the target null values
df_data['PR'] = df_data['PR'].dropna()
df_data
```

```python
cat_col = ['active_snow_tickets', 
           'snow_bin_ticket_minutes',
           'snow_affected_assets',
           'snow_production_level', 
           'low_irradiation', 
           'hurr_bin_ticket_minutes', 
           'hurr_affected_assets', 
           'hurr_production_level', 
           'HurricanePrep', 
           'HurricanePostInspection', 
           'hurricane',
           'storm_active_tickets',
           'storm_bin_ticket_minutes',
           'storm_affected_assets',
           'storm_production_level',
           'lightning',
           'storm',
           'flood',
           'rain']

df_data[cat_col] = df_data[cat_col].astype('category')
```

```python
#Remove unnecessary columns
df_data = df_data.drop(columns=['randid',
                      'storm_affected_assets',
                      'hurr_affected_assets',
                      'snow_affected_assets',
                      'snow_production_level',
                      'hurr_production_level',
                      'storm_production_level',
                      'snow_bin_ticket_minutes',
                      'storm_bin_ticket_minutes',
                      'hurr_bin_ticket_minutes'])
df_data.info()
```
